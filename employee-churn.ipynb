#### Problem Definition

Our objective is to build a classification model that predicts whether a customer will leave (churn) based on their usage patterns, contract type, and account features.

First is importing the necessary libraries


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, accuracy_score
```

Loading and Inspecting the data

```python
df = pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv')

# View first few rows
df.head()
```

```python
print("Shape:", df.shape)
df.info()
df.describe(include='all')
```

Data Cleaning and Preparation

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df = df.dropna(subset=['TotalCharges'])           # Drop rows with missing TotalCharges
df = df.drop_duplicates()
df = df.drop(columns=['customerID'])              # Drop ID column
df['Churn'] = df['Churn'].map({'Yes': 1, 'No': 0})  # Binary encode target
df = pd.get_dummies(df, drop_first=True)          # One-hot encode categorical features
print("Data shape after cleaning:", df.shape)
```

Exploratory Data Analysis (EDA)

```python
sns.countplot(x='Churn', data=df)
plt.title('Churn Distribution')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
corr = df.corr()
sns.heatmap(corr, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
